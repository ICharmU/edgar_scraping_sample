# Gemma  
Note: ollama client is required for GPU acceleration. Otherwise this approach may be impractical. conda does not natively allow for gpu usage

```python
from pydantic import BaseModel, Field
import ollama

class FindYears(BaseModel):
    year: list[int] = Field(ge=1990, le=2030)
    basic: list[float]
    diluted: list[float]

schema = FindYears.model_json_schema()
schema
```

```python
response = ollama.chat(
    model='gemma3:4b',
    format=schema,
    messages=[{
    'role': 'system',
    "content": text,
    },
    {
    'role': 'user',
    'content': f"""
    Find years. 
    Find basic net income PER SHARE.
    Find diluted net income PER SHARE. 
    The values you return are consecutive, by feature.
    The feature name is placed before consecutive values.
    Items are separated by '{splitter}'
    """,
    }],
    options = {
        "temperature": 0,
        "num_gpu": 100,
    }
)


res = response.message.content
res
```

```python
agent_prompt = f"""
    Only find information from 12 months ended. 
    Each X months ended includes current year and previous year.

    Find years. 
    Find basic net income (earnings) PER SHARE.
    Find diluted net income (earnings) PER SHARE. 
    The values you return are consecutive, by feature.
    The feature name is placed before consecutive values.

    Items are separated by '{splitter}'
    """

response = ollama.chat(
    model='gemma3:4b',
    format=schema,
    messages=[{
    'role': 'system',
    "content": text2,
    },
    {
    'role': 'user',
    'content': agent_prompt,
    }],
    options = {
        "temperature": 0,
        "num_gpu": 20,
    }
)


res = response.message.content
res
```
